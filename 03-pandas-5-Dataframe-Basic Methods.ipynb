## Dataframe Methods

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

### apply() method for single column

```python
df.info()
```

```python
#grabbing last 4 digits of the CC number for all rows using user defined function last_four() which will be passed into method apply()

#since cc number is int type, we will convert the formal parameter to str to index it

def last_four(num):
    return int(str(num)[-4:])    
#example: passing 12345678 returns 5678  
#5 is the -4th position, 6 is -3th pos, 7 is -2th pos, 8 is -1th pos

last_four(12345678)
```

```python
#using apply method and create a new columns for the new data

df['CC_last_four'] = df['CC Number'].apply(last_four)
df.head()
```

```python
#determining the expensiveness rate of a restaurant ($, $$ or $$$) using mean and apply methods

df['total_bill'].mean()
```

```python
def expensive(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'

df['Expensiveness_Index'] = df['total_bill'].apply(expensive)
df.head(10)
```

- **Use of lamba expression in place of single line user defined function for apply() method:**

```python
#lambda function

#def last_four(num):
#    return str(num)[-4:]
#
#lamba expression for this function will be:
func = lambda num: str(num)[-4:]
func(12345678)
```

```python
df['CC_last_four'] = df['CC Number'].apply(lambda num: str(num)[-4:])
df.head()
```

### apply() method for multiple columns

```python
#using apply() method to find if a tip is generous or not

def tip_quality(total_bill,tip):
    if tip/total_bill > 0.25:   #tip more than 25% of total bill
        return 'Generous'
    else:
        return 'Other'
    
#pass column names as list and use lambda function in apply()    

df['tip_quality'] = df[['total_bill','tip']].apply(lambda df: tip_quality(df['total_bill'],df['tip']) , axis=1)
df.head(20)

#parameter axis=1 in apply() signifies that total_bill and tip (in apply parameters) are names of columns
```

```python
df[df['tip_quality'] == 'Generous']
```

```python
#another way of writing
#df['tip_quality'] = df[['total_bill','tip']].apply(lambda df: tip_quality(df['total_bill'],df['tip']) , axis=1)

#use np.vectorize()
#more convinient and quicker method

df['tip_quality'] = np.vectorize(tip_quality)(df['total_bill'],df['tip'])
df.head(10)
```
