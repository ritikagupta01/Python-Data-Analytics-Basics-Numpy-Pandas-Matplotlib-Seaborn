# 4. Comparison Plots

Like a 2D version of distribution and categorial plots

- jointplot()
- pairplot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

----
## 4.1 jointplot()

- Can map histograms to each feature of a scatterplot
- Can adjust the scatterplot to be a hex plot or a 2D KDE plot

```python
sns.jointplot(data=df,x='math score',y='reading score',alpha=0.2)
```

```python
# PARAMETER kind=hex

sns.jointplot(data=df,x='math score',y='reading score',kind='hex')
```

```python
# PARAMETER kind=kde

sns.jointplot(data=df,x='math score',y='reading score',kind='kde',
             fill=True)
```

```python
# PARAMETER hue

sns.jointplot(data=df,x='math score',y='reading score',hue='gender')
```

----
## 4.2 pairplot()

- Quick way to compare all numerical columns in a dataframe
- Automatically creates a histogram/KDE for each column and a scatterplot comparison between all possible combinations of columns


```python
sns.pairplot(data=df)
```

```python
# PARAMETER hue

sns.pairplot(data=df,hue='gender',palette='magma')
```

```python
# more parameters

sns.pairplot(data=df,hue='gender',palette='magma',
            diag_kind='hist',corner=True)
```
