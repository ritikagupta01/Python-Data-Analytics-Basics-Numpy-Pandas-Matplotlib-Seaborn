# 5. Seaborn Grids

- Seaborn grid methods use Matplotlib subplots() to automatically create a grid based of a categorial column
- Calling the grid system allows heavy customization of plots


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

## catplot()

```python
# CATPLOT()

# PARAMETERS
# kind: strip, swarm, box, violin, boxen, point, bar, count
# row/col: categorial column to split plots row/col wise

sns.set(style='dark')

sns.catplot(data=df,x='gender',y='math score',kind='boxen',
            row='lunch',col='test preparation course',
           palette='Set1')

plt.savefig('grid.jpg')
```

----
## PairGrid()

```python
# PairGrid()

# creating an empty grid
g = sns.PairGrid(df)

# plotting in the upper triangular plots
g = g.map_upper(sns.scatterplot,color='seagreen',alpha=0.3)

# plotting in the diagonal plots
g = g.map_diag(sns.histplot,lw=0)

# plotting in the lower triangular plots
g = g.map_lower(sns.kdeplot,color='seagreen')
```

```python
# PARAMETER hue
g = sns.PairGrid(df,hue='gender',palette='plasma')

g = g.map_upper(sns.scatterplot,alpha=0.3)
g = g.map_diag(sns.histplot,edgecolor='black')
g = g.map_lower(sns.kdeplot)

g = g.add_legend()
```

----
## FacetGrid()

```python
# FacetGrid()

g = sns.FacetGrid(df,row='lunch',col='gender',hue='test preparation course')

g = g.map(plt.scatter,'math score','reading score',alpha=0.6,edgecolor='white')

plt.legend(bbox_to_anchor=(1.6,1.8))
```

```python
# Adjusting above subplots

g = sns.FacetGrid(df,row='lunch',col='gender',hue='test preparation course')

g = g.map(plt.scatter,'math score','reading score',alpha=0.6,edgecolor='white')

plt.legend(bbox_to_anchor=(1.5,2.6))

# Adjustment
plt.subplots_adjust(hspace=0.2,wspace=0.4)
```
