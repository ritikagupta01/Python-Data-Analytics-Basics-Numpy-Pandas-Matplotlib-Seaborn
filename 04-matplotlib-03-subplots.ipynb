# plt.subplts()

plt.subplots() returns a tuple which by common convention is labeled as **(fig,axes)**

- **fig :** The entire figure canvas
- **axes :** A numpy array holding each of the axes acc to position in the canvas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dataset 1
#exponential relationship

a = np.linspace(0,10,11)
b = a**4
```

```python
# Dataset 2
#linear relationship

x = np.arange(0,10)
y = 2*x
```

```python
# by default, nrows=1, ncols=1
fig,axes = plt.subplots()
```

```python
# returns 1 row and 2 columns
# i.e., two plots side by side
fig,axes = plt.subplots(nrows=1,ncols=2)

# here, axes is a 1D numpy array
```

----

```python
# two rows and two plots in each row
fig,axes = plt.subplots(nrows=2,ncols=2)
```

----
### PLOTTING

```python
# plotting data when axes is one dimensional

fig,axes = plt.subplots(2,1)

# plotting on first plot
axes[0].plot(x,y)
axes[0].set_title('linear data (first plot)')

# plotting on second plot
axes[1].plot(a,b)
axes[1].set_title('exponential data (second plot)')
```

----

```python
# Plotting when axes is two dimensional

fig,axes = plt.subplots(2,2)

axes[(0,0)].plot(x,y)                # 1st way to index in a 2D numpy array
axes[0][0].set_title('first plot')  # 2nd way to index in a 2D numpy array

axes[(0,1)].plot(-x,y)
axes[(0,1)].set_title('second title')

axes[(1,0)].plot(a,b)
axes[(1,0)].set_title('third plot')

axes[(1,1)].plot(-a,b)
axes[(1,1)].set_title('fourth plot')


# FIGURE TITLE
fig.suptitle('Figure Main', fontsize = 20);
```

----

```python
fig,axes = plt.subplots(nrows = 3 , ncols = 1)

axes[0].plot(x,y)

axes[1].plot(a,b)

axes[2].plot(-x,y)
```

```python
# another way to do the above

fig,axes = plt.subplots(nrows = 3 , ncols = 1)

for ax in axes:
    ax.plot(x,y)
```

----
- **Manual spacing**

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

fig.subplots_adjust()

parameters: 

left : position of left edge of subplot, as a fraction of figure width

right : position of right edge of subplots, as a fraction of figure width

bottom : position of bottom edge of subplots, as a fraction of figure height

top : position of top edge of subplots, as a fraction of figure height

wspace : width of padding between subplots, as a fraction of avg Axes width

hspace : height of padding between subplots, as a fraction of avg Axes height


```python
# plotting
fig,axes = plt.subplots(2,2)

axes[(0,0)].plot(x,y)               
axes[0][0].set_title('first plot')  

axes[(0,1)].plot(-x,y)
axes[(0,1)].set_title('second title')

axes[(1,0)].plot(a,b)
axes[(1,0)].set_title('third plot')

axes[(1,1)].plot(-a,b)
axes[(1,1)].set_title('fourth plot')


# MANUAL SPACING

fig.subplots_adjust(wspace=0.5, hspace=0.5)
```

### Exporting the figure

```python
fig,axes = plt.subplots(2,2,dpi=300)

axes[(0,0)].plot(x,y)               
axes[0][0].set_title('first plot')  

axes[(0,1)].plot(-x,y)
axes[(0,1)].set_title('second title')

axes[(1,0)].plot(a,b)
axes[(1,0)].set_title('third plot')

axes[(1,1)].plot(-a,b)
axes[(1,1)].set_title('fourth plot')

# fig.set_dpi(300)
# fig keywords like dpi, figsize etc can be passed in subplots() as well

fig.set_figheight(4)    #inches
fig.set_figwidth(6)

fig.set_facecolor('lightgray')
fig.suptitle('Main Title',fontsize=20)
fig.subplots_adjust(wspace=0.5, hspace=0.5)

# SAVING
fig.savefig('spacing.jpeg',bbox_inches='tight')
```
