### Matplotlib Object Oriented Method

Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object.
<br><br>
The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

```python
import matplotlib.pyplot as plt
```

```python
# creating a blank canvas

plt.figure()
```

```python
# adding axis to a blank canvas

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

# PARAMETERS: [left, bottom, width, height]
# left, bottom, width, height are ratios wrt to the 1 x 1 figure

# left,bottom = 0,0 means axes starts at bottom left corner of blank figure
# width = 1 means width of axes is 100% of the figure width
# height = 1 means height of axes is 100% is the figure height

# axes can be extended outsidethe figure as well
```

```python
# width, height = 0.5, 0.5

plt.figure().add_axes([0,0,0.5,0.5]);
```

```python
# width, height = 0.5, 1

plt.figure().add_axes([0,0,0.5,1]);
```

```python
# width, height = 1, 0.5

plt.figure().add_axes([0,0,1,0.5]);
```

```python
import numpy as np

x = np.arange(0,11)
y = 2*x
```

```python
# plotting on the figure

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y);
```

```python
# one line code for above graph

plt.figure().add_axes([0,0,1,1]).plot(x,y);
```

```python
# FRESH DATA

a = np.linspace(0,10,11)
a
```

```python
b = a**4
b
```

```python
# plotting this new dataset

plt.figure().add_axes([0,0,1,1]).plot(a,b);
```

## MULTIPLE AXES (to make subplots)

**useful for plotting a magnified graph of the same data on the same plot**


```python
fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,10)
axes1.set_ylim(0,10000)
axes1.set_xlabel('x axis')
axes1.set_ylabel('y axis')
axes1.set_title('power of 4')


# Small axes
axes2 = fig.add_axes([0.15,0.4,0.45,0.45])
axes2.plot(a,b)

axes2.set_xlim(6,8)
axes2.set_ylim(0,4000)
axes2.set_title('magnified for x between 6 and 8');
```

```python
# changing positioning of the subplot

fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_title('first plot')

axes2 = fig.add_axes([1,1,1.5,1.5])
axes2.plot(a,b)
axes2.set_title('second plot (150% the size of first)')

axes3 = fig.add_axes([1.1,0.1,1.3,0.7])
axes3.plot(x,y)
axes3.set_title('third plot (width,height = 1.3,0.7)')

axes4 = fig.add_axes([0.1,1.2,0.7,1.2])
axes4.plot(a,b)
axes4.set_title('fourth plot')
```

## Figure Parameters

```python
# PARAMETERS
# dpi : dots per inch 
# figsize : figure size in inches (width x height)

fig = plt.figure(figsize = (8,4) , dpi = 200)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
```
