## 2.  Distribution plots

These plots display a single continuous feature and help visualise properties such as deviation and average values

3 main distribution plot types:
- Rug plot
- Histogram
- KDE plot (kernel density estimation) 
----

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

----
### Rug plot

```python
# rug plot has no y axis
# rug plot is like a lite histogram
# similar to a 1D scatter plot

sns.rugplot(x='salary',data=df,height=0.2)

# the y axis here has no relevance
```

----
### Histogram

- histplot()
- displot()
    - a more generalised distribution plot (hist, kde, rug plots can be plotted together)


```python
# histplot()

plt.figure(figsize=(8,4),dpi=300)
sns.histplot(data=df,x='salary',bins=50,color='red')
```

```python
# displot()

plt.figure(figsize=(3,6),dpi=500)

sns.displot(data=df,x='salary')
```

```python
# view rug plot and kde plot along with histogram in above example

sns.displot(data=df,x='salary',kde=True,rug=True,color='goldenrod')
```

```python
# parameter bin to customise bar/bin width
# method set() to style grid

sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=10,kde=True,
            color='green',edgecolor='black',lw=0.5)
```

```python
# resetting grid style
sns.set(style='white')

# options for style: darkgrid, whitegrid, white, dark, ticks
```

---
### KDE plot
**(Kernel Density Estimation)**

This plot replresents density on its y axis
Plots curve for a normal distribution for the data


```python
sns.set(style='dark')
sns.kdeplot(data=df,x='salary',
            color='purple',ls='--',lw=2)

# resetting background (grid) style
sns.set(style='white')
```

```python
# Parameter clip sets range on the kde plot
# Parameter fill to fill/shade the area under the curve
sns.kdeplot(data=df,x='salary',clip=[60000,120000],fill=True)
```

```python
# Parameter bw_adjust to adjust the bandwidth for multiple normal dist curves (kde plots)

sns.kdeplot(data=df,x='salary',bw_adjust=0.1)
```
