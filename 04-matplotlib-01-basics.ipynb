**Matplotlib is the "grandfather" library of data visualization with Python**

```python
import matplotlib.pyplot as plt

# working with the pyplot portion of the matplotlib library
```

```python
import numpy as np
```

```python
x = np.arange(0,20)
x
```

```python
y = 2*x
y
```

```python
# plotting y vs x graph

plt.plot(x,y)
```

```python
# removing the extra output

plt.plot(x,y)
plt.show()
```

```python
# removing the extra output by adding semicolon

plt.plot(x,y);
```

```python
# customisation

plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xlim(0,10)
plt.ylim(0,20)
plt.title('title');

# saving the image

plt.savefig('firstplot.png')
```
