## 1. Styling - Legends

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# using label parameter to make legend
ax.plot(x,x**2,label='X^2')
ax.plot(x,x,label='X')

# calling legend()
ax.legend()

plt.show()
```

```python
# POSITIONING LEGEND (upper center)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X^2')
ax.plot(x,x,label='X')

ax.legend(loc='upper center')

plt.show()
```

```python
# POSITIONING LEGEND (by coordinates)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X^2')
ax.plot(x,x,label='X')

ax.legend(loc=(0.3,0.5))

plt.show()
```

```python
# POSITIONING LEGEND (by coordinates outside the figure)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X^2')
ax.plot(x,x,label='X')

ax.legend(loc=(1.01,0.9))

plt.show()
```

## 2. Visual Styling

```python
x
```

```python
# LINES

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])


# CHANGING COLOR OF PLOT using color param
# CHANING LINE WIDTH using lw/linewidth param
# CHANGING LINE STYLES using ls/linestyle
  # possible linestyle options: '--', '-', '-.', ':'
    
ax.plot(x,x,color='red',label='x',lw=3,linestyle='--')

ax.plot(x,2*x,color='#cc9600',label='2x',linewidth=3,ls='-.')


# setting custom linestyle
lines = ax.plot(x,3*x + 5,lw=2,label='3x+3 (custom linestyle)')
lines[0].set_dashes([1,2,5,4,15,2])
# parameters: 
# 1 point, 2 white spaces, 5 points, 4 white spaces, 15 points, 2 white spaces

ax.legend();
```

```python
# in the above example for custom linestyling, 
# line is a list containing only one line in this case

# we index this list to customise the line
# since it contains only one line, we index it at the 0 position

print('type(lines) = ',type(lines))
lines
```

```python
# MARKERS
# marking the points in the dataset

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
    
# marker = 'o' gives circles
ax.plot(x,x,color='red',label='x', marker='o')

# marker = 'v' gives downward facing triangle
ax.plot(x,2*x + 2,color='#0084d6',label='2x+2',marker='v')

# marker = '*' gives stars
# markersize/ms param to customize marker size
ax.plot(x,3*x+6,color='deeppink',lw=3,label='3x+6', marker='*',
        ms=15, markerfacecolor='black', markeredgewidth=3, markeredgecolor='deepskyblue')

ax.legend();
```
