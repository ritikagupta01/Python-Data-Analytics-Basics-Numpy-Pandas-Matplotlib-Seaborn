## Conditional Filtering

Selecting rows based on a condition on a column

- Columns -> Attributed of the data
- Rows -> Instances of the data

```python
import numpy as np
import pandas as pd
```

```python
mydata = {'USA':[1776,328,20.5],'CANADA':[1867,38,1.7],'MEXICO':[1821,126,1.22]}
```

```python
df = pd.DataFrame(mydata, index = ['Year','Pop','GDP']).transpose()
df
```

```python
df.info()
```

```python
df.index
```

```python
df.columns
```

- **Filtering by single condition**

```python
df['Pop']
```

```python
#printing indices with the following condition as true or false

df['Pop']>50
```

```python
#finding rows for which above condition is true

df[df['Pop']>50]
```

```python
#reading a new csv file
df = pd.read_csv('tips.csv')
df.head()
```

```python
#grabbing rows where total bill is more than 40 dollars
df[df['total_bill']>40]
```

```python
#list of rows where sex is male
df[df['sex']=='Male']
```

- **Filtering by multiple condition**

```python
#use AND &
#use OR |
```

```python
df.head()
```

```python
#grabbing rows where customer is male and bill is more than 30 dollars
df[(df['sex']=='Male') & (df['total_bill']>30)]
```

```python
#does not work because and operator is built to compare two boolean values

df[(df['sex']=='Male') and (df['total_bill']>30)]
```

- **Check against multiple possible values**

i.e., check for categories or, grouping by name

```python
df.head()
```

```python
#grabbing rows for saturday or sunday or friday
df[(df['day']=='Sat') | (df['day']=='Sun')]
```

```python
#another (concise) way to do the above
options = ['Sat','Sun']
df[df['day'].isin(options)]
```
