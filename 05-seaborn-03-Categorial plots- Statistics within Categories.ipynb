# 3.1 Categorial Plots- Stats within categories

Aim: Diplaying statistical metrics per category

(Visualization equivalent of a groupby() method)

----

Two main types of plots
- countplot() : Counts number of rows per category
- barplot() : General form o displaying any chosen metric per category

----


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df.info()
```

## 3.1.1 countplot()

```python
df['division'].value_counts()
```

```python
# bars along the x axis

plt.figure(figsize=(12,7),dpi=400)

sns.set(style='darkgrid')

sns.countplot(data=df,y='division',palette='Dark2')
```

```python
# bars along the y axis

plt.figure(figsize=(12,7),dpi=400)

sns.set(style='darkgrid')

sns.countplot(data=df,x='division',palette='Dark2')
```

```python
# PARAMETER hue to categorise further

plt.figure(figsize=(12,7),dpi=400)

sns.countplot(data=df,x='division',palette='Dark2',hue='training level')
```

```python
# Manipulating y axis limit to change the appearance

plt.figure(figsize=(12,7),dpi=400)
sns.countplot(data=df,x='division',palette='Dark2')
plt.ylim(90,350)
```

## 3.1.2 barplot()

```python
# visualisation of mean salary per level ofeducation

plt.figure(figsize=(12,6),dpi=400)

sns.barplot(data=df,x='level of education',y='salary',
            estimator=np.mean,errorbar=None,
            palette='brg')
```

```python
# Parameter errorbar set as standard deviation
# will give a black line, whose length denotes the standard deviation

plt.figure(figsize=(12,6),dpi=400)

sns.barplot(data=df,x='level of education',y='salary',
            estimator=np.mean,errorbar='sd',
            palette='brg',errcolor='black')
```

```python
# Parameter hue

plt.figure(figsize=(12,7),dpi=400)

sns.barplot(data=df,x='level of education',y='salary',hue='training level',
            estimator=np.mean,errorbar='sd',
            palette='tab20b',errcolor='black')

plt.legend(bbox_to_anchor=(1,1))
```
