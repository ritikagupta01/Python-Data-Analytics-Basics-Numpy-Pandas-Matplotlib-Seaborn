# 6. Matrix Plots

- Visual equivalent of displaying a pivot table
- Visualizes all the numeric data in the dataframe 
- For efficiency, all the columns chosen should have the same unit

----

Two types:
- heatmap() : visually displays the distribution of cell values with a color mapping
- clustermap() : same visual as heatmap but first conducts hierarchial clustering to reorganize data into groups

----

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
df
```

```python
df.info()
```

```python
# setting labeled index as country
df = df.set_index('Countries')
df
```

----
## 6.1 heatmap()

```python
sns.heatmap(df)
```

```python
# life expectancy column is not suitable for the above heatmap
# making all the columns as rates
# i.e., dropping life expectancy column for heatmap

sns.heatmap(df.drop('Life expectancy',axis=1))
```

```python
# PARAMETERS
# linewidth: lw
# annot: displays value of each cell
# cmap: sets the color mapping of the heatmap
# center: sets value entered as the centre point of the color map in matplotlib

plt.figure(dpi=300)

sns.heatmap(df.drop('Life expectancy',axis=1),
            lw=0.2,annot=True,
            cmap='cividis',center=20)
```

----
## 6.2 clustermap()

```python
plt.figure(dpi=300)

sns.clustermap(df.drop('Life expectancy',axis=1),
            lw=0.2,annot=True,
            cmap='cividis',center=20)
```

```python
# clustering only the index
# PARAMETER col_cluster=False

plt.figure(dpi=300)

sns.clustermap(df.drop('Life expectancy',axis=1),
            lw=0.2,annot=True,
            cmap='cividis',center=20,
            col_cluster=False)
```

```python
# clustering only the columns
# PARAMETER row_cluster=False

plt.figure(dpi=300)

sns.clustermap(df.drop('Life expectancy',axis=1),
            lw=0.2,annot=True,
            cmap='cividis',center=20,
            row_cluster=False)
```
