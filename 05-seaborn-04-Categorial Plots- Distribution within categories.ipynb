```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

## 3.2.1 boxplot()

Drawback: count of data is not decipherable from the boxplot()

The box plot covers the IQR (Inter Quartile Range).
IQR is the range from Q1 to Q3.
- Q1 is the 25% quartile
- Q2 is the 50% quartile (median)
- Q3 is the 75% quartile

```python
# boxplot of math scores categorized by test preparation column

plt.figure(figsize=(7,4),dpi=300)

sns.boxplot(data=df,y='math score',x='test preparation course',palette='Dark2')
```

```python
# PARAMETER hue to further catgorize the data
# further categorizing the boxplots acc to gender 

plt.figure(figsize=(8,4),dpi=300)

sns.boxplot(data=df,
            y='math score',x='test preparation course',
            hue='gender',
            palette='Dark2')
```

----

## 3.2.2 violinplot()

Drawback: Count of data is not decipherable 

Violin plot is useful for comparing probability distributions.

- Displays the prob density across the data using KDE.
- Like a mirrored KDE plot (i.e., KDE plot on both sides of the axis)
- The small black box denoted the IQR and the white dot denotes the median 

```python
plt.figure(figsize=(10,8))

sns.violinplot(data=df,x='math score',y='parental level of education')
```

```python
# PARAMETER hue

plt.figure(figsize=(10,6))

sns.violinplot(data=df,x='math score',y='test preparation course',
               hue='gender',
               palette='Blues')

plt.legend(bbox_to_anchor=(0.155,0.15))
```

```python
# PARAMETER split to split the curves in the above example

plt.figure(figsize=(10,5))

sns.violinplot(data=df,x='math score',y='test preparation course',
               hue='gender',split=True,
               palette='Blues')

plt.legend(bbox_to_anchor=(0.155,0.15))
```

```python
# more parameters in the above example

plt.figure(figsize=(10,5))

sns.violinplot(data=df,x='math score',y='test preparation course',
               hue='gender',split=True,
               inner='quartile',bw=0.1,
               palette='Blues')

plt.legend(bbox_to_anchor=(0.155,0.15))
```

## 3.2.3 swarmplot()

Benefit: Count if data is decipherable 

Shows all the data points in the distribution

For very large datasets, it will display the general distribution of the data

```python
plt.figure(figsize=(6,3.5),dpi=500)

sns.swarmplot(data=df,x='math score',y='gender',size=3,
              hue='test preparation course',
              palette='seismic')

plt.legend(fontsize='x-small',bbox_to_anchor=(0.21,0.16))
```

```python
# PARAMETER dodge in above example

plt.figure(figsize=(6,3.5),dpi=500)

sns.swarmplot(data=df,x='math score',y='gender',size=3,
              hue='test preparation course',dodge=True,
              palette='seismic')

plt.legend(fontsize='x-small',bbox_to_anchor=(0.21,0.16))
```

## 3.2.4 boxenplot() - Letter Value plot

Like a boxplot but with quantiles outside the IQR

```python
sns.boxenplot(data=df,x='math score',y='gender')
```
